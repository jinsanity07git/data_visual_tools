```python
import plotly.plotly as py
from plotly.grid_objs import Grid, Column
from plotly.tools import FigureFactory as FF 

import pandas as pd
import time

url = 'https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv'
dataset = pd.read_csv(url)

table = FF.create_table(dataset.head(10))
```

```python
py.iplot(table, filename='animations-gapminder-data-preview')
```

```python
from pyecharts import Map

value = [155, 10, 66, 78]
attr = ["福建", "山东", "北京", "上海"]
map = Map("全国地图示例", width=1200, height=600)
map.add("", attr, value, maptype='china',is_visualmap=True, )
map
```

```python
from pyecharts import Map

value = [20, 190]
attr = [ '佛堂镇', '金华市']
map = Map("广东地图示例", width=1200, height=600)
map.add("", attr, value, maptype='义乌市', is_visualmap=True,is_label_show=True,
        visual_text_color='#000')
map
```

```python
value = [155, 10, 66, 78]
attr = ["福建", "山东", "北京", "上海"]
map = Map("全国地图示例", width=1200, height=600)
map.add("", attr, value, maptype='china',
        is_visualmap=True, is_piecewise=True,
        visual_text_color="#000",
        visual_range_text=["", ""],
        pieces=[
            {"max": 160, "min": 70, "label": "高数值"},
            {"max": 69, "min": 0, "label": "低数值"},
            {"max": 100, "min": 40, "label": "med数值"},
        ])
map
```
